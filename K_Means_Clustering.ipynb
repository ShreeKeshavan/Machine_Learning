```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
df = sns.load_dataset('iris')
print(df)
```

```python
print(df.columns)
```

```python
df.drop(['sepal_width', 'petal_width','species'],axis = 1, inplace=True)
print(df)
```

```python
#prediction
kmeans = KMeans(n_clusters=3) #n_cluster is to assign number of clusters/groups
kmeans.fit(df)
groups = kmeans.predict(df)
print(groups)
```

```python
df['group'] = groups
sns.scatterplot(data=df,x ='sepal_length', y ='petal_length',hue='group')
plt.show()
```

```python
#from yellowbrick.cluster import KElbowVisualizer
# to known how many groups/clusters need to be formed
kev = KElbowVisualizer(kmeans)
kev.fit(df)
kev.show()
```
